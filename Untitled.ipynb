```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Sample dataset
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace', 'Hannah'],
    'Age': [18, 17, 19, 16, 18, 17, 18, 16],
    'Math': [85, 78, 92, 65, 88, 75, 80, 70],
    'English': [90, 82, 88, 78, 92, 85, 80, 75]
}

# Create a DataFrame from the data
df = pd.DataFrame(data)
```

```python
print(type(data))
print(type(df))
```

```python
# Scatter plot: Age vs. Math score
plt.figure(figsize=(8, 6))
plt.scatter(df['Age'], df['Math'])
plt.xlabel('Age')
plt.ylabel('Math Score')
plt.title('Age vs. Math Score')
plt.show()
```

```python
# Line plot: Age vs. English score
plt.figure(figsize=(8, 6))
plt.plot(df['Age'], df['English'], marker='o')
plt.xlabel('Age')
plt.ylabel('English Score')
plt.title('Age vs. English Score')
plt.grid(True)
plt.show()
```

```python
# Bar plot: Average scores for Math and English
average_scores = df[['Math', 'English']].mean()
plt.figure(figsize=(8, 6))
average_scores.plot(kind='bar')
plt.xlabel('Subject')
plt.ylabel('Average Score')
plt.title('Average Scores for Math and English')
plt.show()
```

```python
# Box plot: Distribution of Math and English scores
plt.figure(figsize=(8, 6))
sns.boxplot(data=df[['Math', 'English']], orient='v')
plt.xlabel('Subject')
plt.ylabel('Score')
plt.title('Distribution of Math and English Scores')
plt.show()
```

```python
# Pair plot: Scatter plot matrix for all numerical columns
sns.pairplot(df[['Age', 'Math', 'English']])
plt.show()
```
