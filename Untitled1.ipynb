```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample dataset
data = {
    'Product': ['Apple', 'Banana', 'Orange', 'Grapes', 'Mango', 'Watermelon', 'Pineapple'],
    'Quantity_Sold': [100, 80, 120, 70, 90, 50, 60],
    'Price': [1.2, 0.8, 1.0, 2.0, 1.5, 4.0, 2.5]
}

# Create a DataFrame from the data
df = pd.DataFrame(data)
```

```python
# Bar plot: Sales quantity of products
plt.figure(figsize=(8, 6))
sns.barplot(x='Product', y='Quantity_Sold', data=df)
plt.xlabel('Product')
plt.ylabel('Quantity Sold')
plt.title('Sales Quantity of Products')
plt.xticks(rotation=45)
plt.show()
```

In this bar plot, each bar represents a product, and the height of the bar corresponds to the quantity sold for that product. We can easily compare the sales quantities of different products. For example, oranges have the highest sales quantity, while watermelons have the lowest.

```python
# Pie chart: Proportion of sales quantity by product
plt.figure(figsize=(8, 8))
plt.pie(df['Quantity_Sold'], labels=df['Product'], autopct='%1.1f%%', shadow=True)
plt.title('Proportion of Sales Quantity by Product')
plt.show()
```

The pie chart represents the proportion of sales quantity for each product as a percentage of the total. This visualization allows us to see the relative contribution of each product to the overall sales. In this case, oranges have the highest share of sales, comprising about 21.7% of total sales.

```python
# Histogram: Distribution of product prices
plt.figure(figsize=(8, 6))
plt.hist(df['Price'], bins=5, edgecolor='black')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Distribution of Product Prices')
plt.show()
```

The histogram displays the distribution of product prices. The x-axis represents price bins, and the y-axis shows the frequency (number of products) falling within each price bin. We can observe that most products are priced around $1.0, and there are fewer products at higher price ranges.

```python
# Box plot: Price distribution for each product
plt.figure(figsize=(8, 6))
sns.boxplot(x='Product', y='Price', data=df)
plt.xlabel('Product')
plt.ylabel('Price')
plt.title('Price Distribution for Each Product')
plt.xticks(rotation=45)
plt.show()
```

The box plot shows the distribution of prices for each product. Each box represents the interquartile range (IQR) of the prices, with the median (50th percentile) shown as a line within the box. The whiskers extend to the minimum and maximum values within 1.5 times the IQR. Outliers are plotted as individual points beyond the whiskers. The box plot allows us to compare the price distributions among different products and identify potential outliers. In this case, "Watermelon" and "Pineapple" have higher prices than other products.

```python
# Scatter plot: Price vs. Sales Quantity
plt.figure(figsize=(8, 6))
plt.scatter(df['Price'], df['Quantity_Sold'])
plt.xlabel('Price')
plt.ylabel('Quantity Sold')
plt.title('Price vs. Sales Quantity')
plt.show()
```

The scatter plot shows the relationship between product prices and sales quantities. Each point represents a product, with the x-coordinate indicating the price and the y-coordinate indicating the quantity sold. This visualization helps us identify any patterns or correlations between price and sales quantity. In this case, there doesn't seem to be a strong linear relationship between price and sales quantity.

```python
# Heatmap: Correlation between sales quantity and price
correlation_matrix = df[['Quantity_Sold', 'Price']].corr()
plt.figure(figsize=(6, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation between Sales Quantity and Price')
plt.show()
```

The heatmap displays the correlation between sales quantity and price. The values in the heatmap represent the correlation coefficient between the two variables. A positive value indicates a positive correlation (as one variable increases, the other tends to increase), while a negative value indicates a negative correlation (as one variable increases, the other tends to decrease). In this case, there is a weak positive correlation (0.11) between sales quantity and price, suggesting that higher-priced products may have slightly higher sales quantities.
